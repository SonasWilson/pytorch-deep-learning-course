###Matrix multiplication

Two types of multiplication:
1. element wise
2. matrix multiplication (dot product)

Rules

1. Inner dimensions should match

* `(3,2) @ (3,2)` not valid
* `(2,3) @ (3,2)` is valid

2. Also the result will be outer dimensions
* `(2,3) @ (3,2) -> (2,2) matrix `

```python
# element-wise
tensor = torch.tensor([1,2,3])
tensor*tensor
```

```python
# matrix multiplication
torch.matmul(tensor, tensor)
```

```python
tensor @ tensor
```

```python
tensor
```

```python
1*1 + 2*2 + 3*3
```

```python
%%time
value = 0
for i in range(len(tensor)):
  value += tensor[i]*tensor[i]
print(value)
```

```python
%%time
torch.matmul(tensor, tensor) #very less time
```

### Reshaping, stacking, squeezing and unsqueezing

* Reshaping - reshapes
* View - view a tensor in a defined shape bt keep the orginal tensor in memory
* Stack - vstack, hstack
* Squeeze - removes all `1` dimensions
* Unsqueeze - add a `1` dimension to target sensor
* Permute - return a view with dimensions swapped

```python
# creating a tensor x
x = torch.arange(1,10)
x, x.shape
```

```python
# reshape (the product should match the size of x)
x_reshaped = x.reshape(1,9) #added a dimension
x_reshaped, x_reshaped.shape
```

```python
# view
z = x.view(1,9)
x, z, z.shape
```

```python
z[:,0] = 5
z, x #changing z will change x as a view of tensor shares the same memory
```

```python
# stacking on top
x_stacked = torch.stack([x, x, x,x,], dim=0)
x_stacked
```

```python
x_reshaped, x_reshaped.shape
```

```python
x_reshaped.shape
```

```python

# squeeze removes all 1
x_squeeze = torch.squeeze(x_reshaped)
x_squeeze, x_squeeze.shape
```

```python
print(f"Previous x_reshaped: {x_reshaped}")
print(f"Previous x_reshaped size: {x_reshaped.shape}")
print("---------------------------------")
print(f"After squeezing x_reshaped {x_squeeze}")
print(f"After squeezing x_reshaped: {x_squeeze.shape}")
```

```python
# torch.unsqueeze() -> adds a single dimension
print(f"Previous tensor: {x_squeeze}")
print(f"Previous size: {x_squeeze.shape}")
print("-----------------------------")
x_unsqueezed = x_squeeze.unsqueeze(dim=0)
print(f"After unsqueezing: {x_unsqueezed}")
print(f"Unaqueezed size: {x_unsqueezed.shape}")
```

```python
# torch.permute() -> rearrange by swapping dimension order
x_original = torch.rand(size=(224,220,3)) #height, width, color channel
x_original.shape
```

```python
# Permuting to  the order width, height, colorchannel
x_permuted = torch.permute(x_original, dims=(1,0,2))
x_permuted.shape
```

```python
print(f"Previous shape: {x_original.shape}")
print(f"After permuting: {x_permuted.shape}")
```

```python
x_original[0,0,0], x_permuted[0,0,0]
```

###Reproducibility

**Note:** A random seed is a number that initializes a pseudorandom number generator (PRNG), which produces a sequence of numbers that appears random but is actually deterministic. Using the same seed will always produce the exact same sequence of numbers, which is useful for testing, debugging, and ensuring reproducibility in applications like video games or scientific simulations.

* `start with random numbers -> tensor operations -> try to make better (again and again and again)`





```python
# creating two random tensors

random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# lets make random but reproducible tensors

# set random seed and call for each tensor
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3,4)
print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
